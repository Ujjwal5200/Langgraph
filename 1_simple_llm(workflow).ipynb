```python
from langgraph.graph  import StateGraph ,START, END
import os
import google.generativeai as genai # Importing the necessary modules
from typing import TypedDict, Dict

from dotenv import load_dotenv

```

```python
# Load environment variables from .env file
load_dotenv()
```

```python
# getting the google api key from the environment variable
#
google_api_key = os.getenv("google_api_key")
genai.configure(api_key=google_api_key)
# calling the generative model

model=genai.GenerativeModel("gemini-2.5-flash")
```

```python
#create a state graph

class llmstate(TypedDict):
    question: str
    answer: str
```

```python
def llm_qa(state):
    question = state["question"]
    prompt = f" Answer the following question precisely { question}"
    model = genai.GenerativeModel("gemini-2.5-flash")
    response = model.generate_content(prompt)
    state["answer"] = response.text
    return state
```

```python
#create our graph
graph = StateGraph(llmstate)
# add nodes to the graph
graph.add_node(
    "llm_qa",llm_qa)
# add the edges to the graph
graph.add_edge(START, 'llm_qa')
graph.add_edge('llm_qa', END)
#compile the graph
workflow=graph.compile()
```

```python
initial_state=({"question": "What is the capital of usa?"})
final_state = workflow.invoke(initial_state)
print(final_state)
# final_state will contain the answer to the question
# you can access the answer like this
# this will return the answer in the state
```
