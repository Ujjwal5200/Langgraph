```python
from langgraph.graph  import StateGraph ,START, END
import os
import google.generativeai as genai # Importing the necessary modules
from typing import TypedDict, Dict

from dotenv import load_dotenv
```

```python
# Load environment variables from .env file
load_dotenv()
```

```python
#getting the API key from environment variables
google_api_key = os.getenv("google_api_key")
genai.configure(api_key=google_api_key) # Configuring the API key for Google Generative AI
# calling the model (gemini-2.5-flash) to generate text
model = genai.GenerativeModel("gemini-2.5-flash") # Initializing the
```

```python
class blogstate(TypedDict):
    title: str
    content: str
    outline: str
    evaluation: str
```

```python
def create_outline(state: blogstate) -> blogstate:
    """
    Create an outline for a blog post based on the title.
    """
    prompt = f"Create an outline for a blog post titled '{state['title']}'"
    response = model.generate_content(prompt)
    state['outline'] = response.text.strip()
    return state
```

```python
def write_content(state: blogstate) -> blogstate:
    """
    Write the content of the blog post based on the outline.
    """
    prompt = f"Write a blog post based on the following outline:\n{state['outline']}"
    response = model.generate_content(prompt)
    state['content'] = response.text.strip()
    return state
```

```python
def evaluate_content(state: blogstate) -> blogstate:
    """
    Evaluate the content of the blog post.
    """
    prompt = f"Evaluate the following blog post content on the base of my outine out of 10 :\n{state['content']}"
    response = model.generate_content(prompt)
    state['evaluation'] = response.text.strip()
    return state
```

```python
# starting the state graph
graph = StateGraph(blogstate)
# nodes
graph.add_node('create_outline',create_outline)
graph.add_node('write_content',write_content)
graph.add_node('evaluate',evaluate_content) # Assuming evaluate is defined elsewhere
# edges
graph.add_edge(START, 'create_outline')
graph.add_edge('create_outline', 'write_content') 
graph.add_edge('write_content', 'evaluate')
graph.add_edge('evaluate', END)  
workflow=graph.compile()
```

```python
initial_state=({'title': 'The Future of AI in Everyday Life',
 'content': '',
 'outline': '',
 'evaluation': ''})  # tell about the evaluation of the content
final_state = workflow.invoke(initial_state)
print(final_state)
print(final_state['evaluation'])  # Print the evaluation of the content
```
